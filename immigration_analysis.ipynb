```python
cols_to_drop = ['Type','Coverage','AREA','REG','DEV']
df = df.drop(columns= cols_to_drop)
```

```python
df= df.rename(columns={
    'OdName': 'country',
    'AreaName': 'continent',
    'RegName': 'region',
    'DevName': 'status'
})
```

```python
df.sort_values(by=2013,ascending=False)
```

add column

```python
df[1980]+df[1981]
```

```python
years= range(1980,2014)
df[years].sum(axis=1)
```

```python
df['total']=df[years].sum(axis=1)
```

```python
df
```

replacing value

```python
df['status'] = df['status'].replace('Developing regions',False)
df['status'] = df['status'].replace('Developed regions',True)
df = df.rename(columns={'status':'developed'})
```

```python
df.info()
```

saving the processed data

```python
df.to_csv('dataset/canada_clean.csv', index = False)
```
